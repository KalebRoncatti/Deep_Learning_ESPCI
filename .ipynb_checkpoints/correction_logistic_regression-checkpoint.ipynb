```python
import sys
import numpy as np 
from numpy.linalg import norm
from numpy.random import shuffle
import matplotlib.pyplot as plt
from toydata import * 
from numpy.linalg import norm  # norm of a vector
%matplotlib inline  
%config InlineBackend.figure_formats=['png']
%config InlineBackend.figure_format = 'png'
```

```python
X = Xstudents
Y = Ystudents
print(X.shape)

plot2DBinaryDataSet(X[:,Y==0],X[:,Y==1],xmin=-1,xmax=21,ymin=-1,ymax=21)
# shortcuts
def plotDataStudents():
    plot2DBinaryDataSet(X[:,Y==0],X[:,Y==1],xmin=-1,xmax=21,ymin=-1,ymax=21)
def plotLineStudents(w,b):
    plotLine(w,b,xmin=-1,xmax=21,ymin=-1,ymax=21)
plotDataStudents()
```

```python
def init_weights(std=0.1): 
    """Initialize weights with a gaussian of 0 mean
    Args: @
        - std: the variance 
    Returns: 
        - w is a np.array of shape (1,2) 
        - b is a float
    """
    w = np.random.randn(1,2)*std
    b = np.random.randn()*std
    return w,b

w,b = init_weights()
print(w,b)

def sigmoid(x):
    """
    Sigmoid function for numpy array (as input and and output)
    """
    return  1/(1 + np.exp(-x))


def inference(x,weights,bias):
    """ 
    Logist regression inference 
    
    Inputs: 
        - x : np array (d,N), 
            with d the dimension on an example 
            and N the number or examples
            
        - weights: np.array of dimension (1,d)
        - bias term : a float
        
    Outputs: 
    sigmoid(weights@x+bias)
    An np array of size (1,N) 
    """
    return sigmoid(weights@x+bias)
# minimal test
x = X[:,0]
y = Y[0]
h = inference(x,w,b)
print(x)
print(h)
```

```python
def loss_fn(y,h): 
    return -y*np.log(h) - (1-y)*np.log(1-h)
    
def gradient_loss(y,h,x):
    grad = (h-y)
    return (grad*x, grad)


dw , db = gradient_loss(y,h,x)
print(dw, db)
print(loss_fn(y,h))
```

# Online training 

Basic experiment

```python
eta = 1e-2   # Learning Rate
Nepoch = 1000
Nprint = Nepoch/10 # 10 prints
# init. 
w,b = init_weights()
print("Init. value (random):" ,  w , b)
loss_list=[] # to store loss values
err_list=[]  # to store the error rate 

# for the randomized order
idx = np.arange(len(Y))
print("epoch\tloss\t%error")
for e in range(Nepoch): 
    epochloss = 0
    errorRate = 0 
    np.random.shuffle(idx)
    for i in idx:
        h = inference(X[:,i],w,b).squeeze()
        d = gradient_loss(Y[i],h,X[:,i])
        epochloss += loss_fn(Y[i],h)
        w = w - eta*d[0]
        b = b - eta*d[1]
        hyp = 0
        if h > 0.5: 
            hyp = 1  
        if hyp != Y[i]:
            errorRate +=1
    errorRate = errorRate*100.0/len(Y)
    err_list.append(errorRate)
    loss_list.append(epochloss)
    
    if e%Nprint == 0 :
        print(e,"\t{0:.2f}".format(epochloss),"\t{0:.2f}\t".format(errorRate))
        
```

Look at the training progression 


```python
Xup = 1000 #len(loss_list) to see everything, less to zoom in
plt.plot(range(Xup),loss_list[:Xup],"b-")
plt.show()

plt.plot(range(Xup),err_list[:Xup],"r-")
```

After the iteration 300, the loss is still decreasing while the classification error rate does not move anymore

Look at the norm of the parameters 

```python
eta = 1e-2
Nepoch = 10000
Nprint = Nepoch/10 # 10 prints
# init. 
w,b = init_weights()
print("Init. value (random):" ,  w , b)
loss_list=[] # to store loss values
nw_list=[]   # to store the weight norm 
err_list=[]  # to store the error rate 
# for the randomized order
idx = np.arange(len(Y))
for e in range(Nepoch): 
    epochloss = 0
    errorRate = 0 
    np.random.shuffle(idx)
    for i in idx:
        h = inference(X[:,i],w,b).squeeze()
        d = gradient_loss(Y[i],h,X[:,i])
        epochloss += loss_fn(Y[i],h)
        w = w - eta*d[0]
        b = b - eta*d[1]
        nw = norm(w)**2 + b**2
        hyp = 0
        if h > 0.5: hyp = 1  
        if hyp != Y[i]:
            errorRate +=1
    errorRate = errorRate*100.0/len(Y)
    err_list.append(errorRate)
    loss_list.append(epochloss)
    nw_list.append(nw)
    
    if e%Nprint == 0 :
        print(e,"\t{0:.2f}".format(epochloss),"\t{0:.2f}\t".format(errorRate),nw,w/nw,b/nw)
        #print(e,"\t",epochloss,nw,w/nw,b/nw)
        #plotDataStudents()
        #plotLineStudents(w.T, b)
        #plt.show()
    
```

```python
Xup = 10000 #len(loss_list) to see everything, less to zoom in
plt.plot(range(Xup),loss_list[:Xup],"b-")
plt.show()
plt.plot(range(Xup),err_list[:Xup],"r-")
plt.show()

plt.plot(range(Xup),nw_list[:Xup],"g-")
plt.show()

```

Batch training

```python
eta = 1e-3
Nepoch = 1000
Nprint = Nepoch/10 # 10 prints
# init. 
w,b = init_weights()
loss_list=[]
nw_list=[]
for e in range(Nepoch): 
    H = inference(X,w,b).squeeze()
    D = gradient_loss(Y,H,X)
    epochloss = loss_fn(Y,H).sum()
    w = w - eta*D[0].mean(axis=1)
    b = b - eta*D[1].mean()
    nw = (norm(w)+b**2).squeeze()
    if e%Nprint == 0 :
        print(e,epochloss,nw)
        #plotDataStudents()
        #plotLineStudents(w.T, b)
        #plt.show()
    loss_list.append(epochloss)
    nw_list.append(nw)
    
```

```python
Xup = 1000 #len(loss_list) to see everything, less to zoom in
plt.plot(range(Xup),loss_list[:Xup],"b-")
plt.show()
plt.plot(range(Xup),err_list[:Xup],"r-")
plt.show()

plt.plot(range(Xup),nw_list[:Xup],"g-")
plt.show()

```

# NL dataset 1 

```python
import pickle 
data1,data2 = pickle.load( open('nl_dataset1.pck', 'rb'))
```

```python
# Make a dataset 
X = np.concatenate((data1,data2), axis=1)
print(X.shape)
Y = np.ones(data1.shape[1]+data2.shape[1])
Y[:data1.shape[1]]=0
print(Y.shape,Y.sum())
```

```python
minvalues = X.min(axis=1)
maxvalues = X.max(axis=1)
XLIM = np.array([minvalues[0], maxvalues[0]])
YLIM = np.array([minvalues[1], maxvalues[1]])

print(XLIM,YLIM)
```

```python
# nicer like this 
plt.plot(data1[0,:], data1[1,:],'r.',alpha=0.5)
plt.plot(data2[0,:], data2[1,:],'g.',alpha=0.5)
plt.xlim(XLIM)
plt.ylim(YLIM)
plt.grid(True)
```

```python
eta = 1e-3
Nepoch = 1000
Nprint = Nepoch/10 # 10 prints
# init. 
w,b = init_weights()
print(w,b)
loss_list=[]
nw_list=[]
for e in range(Nepoch): 
    H = inference(X,w,b).squeeze()
    D = gradient_loss(Y,H,X)
    epochloss = loss_fn(Y,H).sum()
    w = w - eta*D[0].mean(axis=1)
    b = b - eta*D[1].mean()
    nw = np.sqrt(w@w.T+b**2).squeeze() 
    if e%Nprint == 0 :
        print(e,epochloss,nw,w/nw,b/nw)
    loss_list.append(epochloss)
    nw_list.append(nw)
    
print(w,b)
```

```python
# nicer like this 
plt.plot(data1[0,:], data1[1,:],'ro',alpha=0.5)
plt.plot(data2[0,:], data2[1,:],'go',alpha=0.5)
plotLine(w.T,b,XLIM[0],XLIM[1],YLIM[0],YLIM[1])
print(w,b)
plt.show()
```
