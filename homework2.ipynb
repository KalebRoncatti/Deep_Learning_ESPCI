# - Implement a feedforward NNet with one hidden layer in pytorch with the Sequential module (the hidden layer size is 2). This is an extension of the logistic regression model. 

# - Write the code to train the model and run it (be sure to monitor the loss function during the training process)

```python
import torch as th
import time
start_time = time.time()

#Creating the model and training the network

D_in=2  # input in the 1st layer
D_out=1 # output 
D_h=2  # input of the hidden layer

# Set the desired model, in our case, we have 2 neurons in the input, the sigmoid function will be applied and result in 
# 2 outputs. Our hidden layer will take these outputs as new inputs and reapply the sigmoide, but in this case it will 
# result in 1 output.
model = th.nn.Sequential(th.nn.Linear(D_in, D_h),th.nn.Sigmoid(),th.nn.Linear(D_h,D_out),th.nn.Sigmoid())

print(model)

# Applying the loss function
loss_fn = th.nn.BCELoss() # The binary cross entropy 
learning_rate = 1e-1
# Applying the optimization function
optimizer = th.optim.SGD(model.parameters(), lr=learning_rate)


Nepochs = 1000 # number of epochs 
Nprint  = Nepochs/10  # frequence of print

totalloss_list=[]
err_list=[]

TOT = np.arange(len(Y))

print('#########################################')
print("epoch\tloss\t%error")
for epoch in range(Nepochs):
    total=0
    errorRate = 0 
    for i in TOT:
        prediction = model(X[i]).squeeze()    
        loss = loss_fn(prediction, Y[i])
        optimizer.zero_grad()  
        loss.backward()        
        optimizer.step()
        total+=loss
        hyp = 0
        if prediction > 0.5: 
            hyp = 1  
        if hyp != Y[i]:
            errorRate +=1
    
    errorRate = errorRate*100.0/len(Y)
    totalloss_list.append(total)
    err_list.append(errorRate)
    if epoch%Nprint == 0:
        print(epoch,"\t{0:.2f}".format(total),"\t{0:.2f}\t".format(errorRate))
        
print("--- %s seconds ---" % (time.time() - start_time))

```

# - Plot the learning curve (the evolution of the loss function w.r.t the epoch)

```python
# Compute the learning curve along the epochs and the error rate

Xe = len(totalloss_list)
plt.plot(range(Xe),totalloss_list[:Xe],"b-")
plt.show()
plt.plot(range(Xe),err_list[:Xe],"r-")
plt.show()
```

# - Shortcuts

```python
# Shortcuts

#Function to plot a line through parameters w (u) and bias

def plotLine(u,b,xmin, xmax, ymin, ymax, linestyle = "-", color="k" ):
    """Plot a line in the "box" defined by xmin, xmax, ymin, ymax
    
    Args: 
        - u : The line is defined by its normal vector u, a numpy array of size 2
        - b : the constant term
        - the "box" is defined by xmin, xmax, ymin, ymax
        
    The line has the equation:     u[0]*x + u[1]*y + b = 0 
    """
    if u[1]==0:
        x=-b/u[0]
        plt.plot([x,x],[ymin,ymax])
    else:
        y1=(-b-u[0]*xmin)/u[1]
        y2=(-b-u[0]*xmax)/u[1]
        plt.plot([xmin,xmax],[y1,y2],color='k')
        plt.axis([xmin,xmax,ymin,ymax])
        
```

# - Compute the tensor Z that represents the projection of all the inputs (X) in the latent representation space (the associated values of the hidden layer). To do that, you can get a reference of each module and especially the input (Linear) layer (see idl_pytorch_starter notebook) and use the input layer to transform X in Z with the forward operation

```python

# Here we are subsequently:
# 1. applying the first model (linear - model[0]) in X.
# 2. applying the sigmoide function (model[1]) in the result obtained previously
# 3. We are going to obtain HOW the points x,y are transformed along these steps (1. and 2.)
mod=model[1](model[0](X))

# We visualize the form and compute its shape
print(mod.shape)

# We use detach because this way we are able to remove the gradient and transform the tensor in a numpy() array
mod=mod.detach()
mod=mod.numpy()

# Here we will be taking the bias and Weigth generated by the model after the application of the sigmoide function
newmod=model[2]
Bhidden=newmod.bias.data[0]
Whidden=newmod.weight.data[0]


shape1=data1.shape
shape=Xbefore.shape

minvaluesH = mod.min(axis=0)
maxvaluesH = mod.max(axis=0)

# Plotting the points separated in classes as before

hiddenXred=mod[0:shape1[1],0]
hiddenYred=mod[0:shape1[1],1]
hiddenXgreen=mod[shape1[1]+1:shape[1],0]
hiddenYgreen=mod[shape1[1]+1:shape[1],1]
plt.plot(hiddenXred, hiddenYred,'b.',alpha=0.7)
plt.plot(hiddenXgreen, hiddenYgreen,'r.',alpha=0.7)

# Applying the function plot for the line in the new space

plotLine(Whidden,Bhidden,-.2+minvaluesH[0],maxvaluesH[0]+.2,-.2+minvaluesH[1],maxvaluesH[1]+.2,linestyle="-",color="k")
plt.show()
```

# - Plot the original dataset X along with the two linear separations found by the hidden layer

```python



# What the method does is to select RANDOM values of w and b, and after each backpropagation, these values w and b are better
# chosen to separate the dataset. Here, we will be plotting the TWO lines that we found in the original dataset along with the 
#TWO linear separations were settled by the hidden layer

mod=model[0]
B0=mod.bias.data[0]
B1=mod.bias.data[1]
W0=mod.weight.data[0]
W1=mod.weight.data[1]

minvalues = Xbefore.min(axis=1)
maxvalues = Xbefore.max(axis=1)
#print(minvalues)
#print(maxvalues)
XLIM = np.array([minvalues[0], maxvalues[0]])
YLIM = np.array([minvalues[1], maxvalues[1]])
plt.plot(data1[0,:], data1[1,:],'b.',alpha=0.7)
plt.plot(data2[0,:], data2[1,:],'r.',alpha=0.7)
plt.xlim(XLIM)
plt.ylim(YLIM)
plt.grid(True) 
    
plotLine(W0,B0,minvalues[0],maxvalues[0],minvalues[1],maxvalues[1],linestyle="-",color="k")
plotLine(W1,B1,minvalues[0],maxvalues[0],minvalues[1],maxvalues[1],linestyle="-",color="k")
plt.show()
```
